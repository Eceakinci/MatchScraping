```python
if 112 & 30 in df.index:
    original_row_113 = df.loc[112]
    original_row_30 = df.loc[30]

    print(f"  match_id: {original_row_113.get('match_id', 'N/A') == original_row_30.get('match_id', 'N/A')}")
    print(f"  name: {original_row_113.get('name', 'N/A') == original_row_30.get('name', 'N/A')}")
    print(f"{original_row_30.get('name', 'N/A')}")
    print(f"{original_row_113.get('name', 'N/A')}")

    print(f"  shirt_number: {original_row_113.get('shirt_number', 'N/A') == original_row_30.get('shirt_number', 'N/A')}")
```

This check is important because when I was trying to combine all data columns based on match ID and player, I was only getting 5 rows for one match, but there should be 6 different types of data per match. When attempting to combine all data for one player and match, the program was creating 2 rows instead of 1. This happens because trailing spaces in the name column cause equality checks to fail.

```python
df['name'] = df['name'].apply(lambda x: x.strip())
df = df.groupby(['match_id', 'name', 'shirt_number']).first().reset_index()
```

```python
df.to_csv("combined_data.csv", sep='\t')
```

```python
attack_df = df[
    (df["match_id"] == 22403) &
    (df["attack_efficiency_percentage"] is not None) &
    df["position"].isin(["OH", "MB", "O"])
]

player_attack_avg = (
    attack_df.groupby("name")["attack_efficiency_percentage"]
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

player_attack_avg.plot(kind="barh", figsize=(8, 5))
plt.title("Top 5 Players by Average Attack Total")
plt.xlabel("Average Attack Total per Match")
plt.show()
print(attack_df)
```

Well, a graph with setters gives us no insight about attack efficiency, because if the setter attacked once, it seems like the best player in terms of attack. That's why we need to compare attack data only for attackers - Opposites, Opposite Hitters, Middle Blockers
